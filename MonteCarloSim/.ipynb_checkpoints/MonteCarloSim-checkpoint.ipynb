```python
vol = []
for i in range(len(deltaY)):
    vol.append(volPct[i] * Y[i])
vol
```

```python
trials = 100
```

```python
paths = []
for i in range(trials):
    sum_arr = []
    rand = []
    for j in range(len(deltaY)):
        rand.append(np.random.lognormal(deltaY[j], vol[j]))
    sum_arr.append(rand[0])
    for k in range(1, len(deltaY)):
        sum_arr.append(rand[k] + rand[k - 1])
    paths.append(sum_arr)
    plt.plot(X, sum_arr)
plt.show()
```

```python
prep_arr = [0.95] * len(X)
```

```python
def prep(arr, p_arr):
    result = [0] * len(arr)
    for i in range(len(arr)):
        result[i] = prep_arr[i] * arr[i] - prep_arr[i] * arr[i] * arr[i]
    return result
```

```python
prep(paths[0], prep_arr)
```

```python
def payment(interest_rate, number_of_payments, principal):
    M = principal*(interest_rate/12 * (1+interest_rate/12)** number_of_payments/(((1+interest_rate/12)**number_of_payments)-1))
    return M
```

```python
payment(0.03, 360, 100000)
```

```python
def npv(irr, cfs, rates, yrs):
    total = 0
    for i in range(yrs):
        total += cfs[i] / ((1 + rates[i] + irr) **i)
    return total
```
